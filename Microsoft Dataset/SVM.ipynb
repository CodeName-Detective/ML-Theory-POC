# Missing Values

```python
missing_values_df = pd.DataFrame({'Column_Name': df.columns,
             'missing_value_pct': (df.isnull().sum()/df.shape[0])*100})
missing_value_remove_columns= missing_values_df[missing_values_df['missing_value_pct']>50]['Column_Name'].values
missing_value_remove_columns
```

```python
df.drop(columns=missing_value_remove_columns,axis=1,inplace=True) # Droping the cloumns
```

```python
for col in df.columns.values:
    if df[col].dtype in ['int64','float64']:
        df[col].fillna(df[col].mean(),inplace=True)
    elif df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0],inplace=True)
```

# Remove Unique and Same valued Columns

```python
for col in df.columns.values:
    if df[col].dtype == 'object':
        if df[col].nunique() == 1 or df[col].nunique() == df.shape[0]:
            df.drop(col,axis=1,inplace=True)
```
