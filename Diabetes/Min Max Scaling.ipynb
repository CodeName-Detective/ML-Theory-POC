# Train Test Split

```python
from sklearn.model_selection import train_test_split
y = df['Outcome']
X = df.drop(columns='Outcome')
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=17)
```

# Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
```

- Here we get min and max values by using __fit()__ function
- Then we do scaling using the __transform()__ function
- both can be done at a time by using __fit_transform()__ function

- Here we get _fit_ using the __train data__ and _transform_ the __train data__
- As far as the __test data__ is considered it is scaled by the scale created over the __train data__ so that training and testing the model is being happend over the same scale i.e., Scale of __train data__

```python
for col in X.columns:
    X_train[col] = min_max_scaler.fit_transform(np.array(X_train[col]).reshape(-1,1)) #Fitting and Transforming over train data
    X_test[col] = min_max_scaler.transform(np.array(X_test[col]).reshape(-1,1)) # Only Transforming over test data
```
