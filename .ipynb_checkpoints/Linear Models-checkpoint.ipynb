```python
import numpy as np
X = 2* np.random.rand(100,1)
y = 4+3*X+np.random.randn(100,1)
```

# The Normal Equation 

```python
X_b = np.c_[np.ones((100,1)),X] #Add x0=1(Bias) to the Features
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
import matplotlib.pyplot as plt
plt.plot(X_new,y_predict,"r-")
plt.plot(X,y,"b.")
plt.axis([0,2,0,15])
plt.show()
```

# Batch Gradient Descent

```python
eta = 0.1
n_iterations = 1000
m =100 #Number of Records

theta = np.random.randn(2,1)
```

```python
for iteration in range(n_iterations):
    gradients = 2/m*X_b.T.dot(X_b.dot(theta)-y) #Gradient Vector
    theta = theta - eta*gradients
```

```python
theta
```

# Stochastic Gradinet Descent

```python
n_epochs = 50
m=100
theta = np.random.randn(2,1)
```

```python
t0,t1 = 5,50 # Learning Schedule Parameter
def learning_schedule(t):
    return t0/(t+t1)
```

```python
for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m) #Select a random interger with in range of m
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta)-yi)
        eta = learning_schedule(epoch*m+i)
        theta = theta-eta*gradients
```

```python
theta
```

# Learning Curves

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
def plot_learning_curves(model,X,y):
    X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.2)
    train_errors,val_errors = [],[]
    for m in range(1,len(X_train)):
        model.fit(X_train[:m],y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m],y_train_predict))
        val_errors.append(mean_squared_error(y_val,y_val_predict))
    plt.plot(np.sqrt(train_errors),"r-+",linewidth=2,label="train")
    plt.plot(np.sqrt(val_errors),"b-",linewidth=3,label="val")
    plt.legend()
```

### For Linear Regression

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg,X,y)
```

### For Polynomial Regression

```python
from sklearn.pipeline import Pipeline
polynomial_regression = Pipeline([
    ('poly_features',PolynomialFeatures(degree=2,include_bias=False)),
    ('lin_reg',LinearRegression())
])
```

```python
plot_learning_curves(polynomial_regression,X,y)
```

# Early Stopping

```python
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.base import clone
from sklearn.preprocessing import StandardScaler
poly_scaler = Pipeline([
    ("poly_features",PolynomialFeatures(degree=90,include_bias=False)),
    ("std_scaler",StandardScaler())
])
```

```python
X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)
```

```python
sgd_reg = SGDRegressor(max_iter=1,tol=-np.infty,warm_start=True,
                      penalty=None,learning_rate='constant',eta0=0.0005)
```

```python
minimum_val_error = float('inf')
best_epoch =None
best_model = None
```

```python
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled,y_train)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val,y_val_predict)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
```

```python
best_epoch, best_model
```

# Logistic Function

```python
from sklearn import datasets
iris = datasets.load_iris()
iris.keys()
```

```python
X = iris['data'][:,3:]
y = (iris['target']==2).astype(np.int)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)
```

```python
plt.plot(X_new,y_proba[:,1],"g-",label="Iris Virginica")
plt.plot(X_new,y_proba[:,0],"b--",label="Not Iris Virginica")
plt.legend()
```

```python
log_reg.predict([[1.7],[1.5]])
```
